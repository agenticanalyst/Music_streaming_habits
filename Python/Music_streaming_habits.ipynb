# Exploratory Data Analysis (EDA)

```python
# Age Distribution
# Platform Distribution
# Subscription Distribution
# Top Genres
# Top Artists
# Country-wise Users
# Daily Listening Minutes
# Songs Per Day
# Skip Rate Analysis
# Mood Distribution
# Offline Mode Usage
# Podcast Users
```

```python
# Age Distribution
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.hist(df['age'], bins=10, edgecolor='black')

plt.title("Age Distribution of Music Listeners")
plt.xlabel("Age")
plt.ylabel("Number of Users")

plt.show()
```

```python
# Platform Distribution
platform = df['platform'].value_counts()

plt.figure(figsize=(8,5))
platform.plot(kind='bar')

plt.title("Platform Distribution")
plt.xlabel("Platform")
plt.ylabel("Number of Users")
plt.xticks(rotation=0)

plt.show()
```

```python
# Subscription Distribution
subscription = df['subscription'].value_counts()

plt.figure(figsize=(6,6))
subscription.plot(kind='pie', autopct='%1.1f%%', startangle=90)

plt.title("Subscription Distribution")
plt.ylabel("")

plt.show()
```

```python
# Top Genres
genre = df['top_genre'].value_counts()

plt.figure(figsize=(10,5))
genre.plot(kind='bar')

plt.title("Top Music Genres")
plt.xlabel("Genre")
plt.ylabel("Number of Users")
plt.xticks(rotation=45)

plt.show()
```

```python
# Top Artists
artist = df['top_artist'].value_counts().head(10)

plt.figure(figsize=(10,5))
artist.plot(kind='bar')

plt.title("Top 10 Artists")
plt.xlabel("Artist")
plt.ylabel("Number of Users")
plt.xticks(rotation=45)

plt.show()
```

```python
# Country-wise Users
country = df['country'].value_counts()

plt.figure(figsize=(10,5))
country.plot(kind='bar')

plt.title("Country-wise Users")
plt.xlabel("Country")
plt.ylabel("Number of Users")
plt.xticks(rotation=45)

plt.show()
```

```python
# Daily Listening Minutes
plt.figure(figsize=(8,5))
plt.hist(df['daily_listening_minutes'], bins=10, edgecolor='black')

plt.title("Daily Listening Minutes")
plt.xlabel("Minutes")
plt.ylabel("Number of Users")

plt.show()
```

```python
# Songs Per Day
plt.figure(figsize=(8,5))
plt.hist(df['songs_per_day'], bins=10, edgecolor='black')

plt.title("Songs Played Per Day")
plt.xlabel("Songs")
plt.ylabel("Number of Users")

plt.show()
```

```python
# Skip Rate Analysis
plt.figure(figsize=(8,5))
plt.hist(df['skip_rate_pct'], bins=10, edgecolor='black')

plt.title("Skip Rate Distribution")
plt.xlabel("Skip Rate (%)")
plt.ylabel("Number of Users")

plt.show()
```

```python
# Mood Distribution
mood = df['top_mood'].value_counts()

plt.figure(figsize=(8,5))
mood.plot(kind='bar')

plt.title("Mood Distribution")
plt.xlabel("Mood")
plt.ylabel("Number of Users")
plt.xticks(rotation=0)

plt.show()
```

```python
# Offline Mode Usage
offline = df['uses_offline_mode'].value_counts()

plt.figure(figsize=(6,6))
offline.plot(kind='pie', autopct='%1.1f%%', startangle=90)

plt.title("Offline Mode Usage")
plt.ylabel("")

plt.show()
```

```python
# Podcast Users
podcast = df['podcasts_too'].value_counts()

plt.figure(figsize=(6,6))
podcast.plot(kind='pie', autopct='%1.1f%%', startangle=90)

plt.title("Podcast Users")
plt.ylabel("")

plt.show()
```

```python
genre_mood = pd.crosstab(df['top_genre'], df['top_mood'])

genre_mood.plot(kind='bar', figsize=(10,5))

plt.title("Genre vs Mood")
plt.xlabel("Genre")
plt.ylabel("Number of Users")
plt.xticks(rotation=45)

plt.show()
```

```python
platform_subscription = pd.crosstab(df['platform'], df['subscription'])

platform_subscription.plot(kind='bar', figsize=(8,5))

plt.title("Platform vs Subscription")
plt.xlabel("Platform")
plt.ylabel("Number of Users")

plt.show()
```
